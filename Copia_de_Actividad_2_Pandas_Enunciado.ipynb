````python
abb = abb[['state','abbreviation']]
area       = area[['state','area (sq. mi)']]
population        = population[['state/region','ages','year','population']]
````

````python
# Pregunta 4
population.groupby(['state/region','ages']).agg(['mean','min','max','count'])
````

````python
# Pregunta 5
df0 = pd.merge(abb,area,on= 'state', how = 'outer')
print(df0.info())
df = pd.merge(population,df0, left_on='state/region',right_on='abbreviation',how='left') #toma todo los datos de la izquierda y le agrega los de la derecha
print(df.info())
df.head()
````

````python
df.isnull().sum()
````

````python
df.tail()
````

````python
# Pregunta 6
aux = df.groupby('state/region')['population'].mean()
print(aux.head())
df_aux =  pd.DataFrame({'state' : aux.index, 'pop/area' : aux.values})
df_aux.head()


````

````python
population.head()
````

````python
area.head()
````

````python
paso1 = pd.merge(abb,df_aux, left_on='abbreviation',right_on='state',how='inner')
paso1.head()
````

````python
paso2 = pd.merge(paso1,area,left_on='state_x',right_on='state',how='inner')
#inner es la opcion por defecto, toma los elementos comunes entre las dos tablas
paso2.head()
````

````python
paso3 = paso2[['abbreviation','state','area (sq. mi)','pop/area']]
paso3.set_index('abbreviation', inplace = True)
paso3
````
